```python
# Import necessary libraries
import pandas as pd
import os
from sqlalchemy import create_engine

# Paths to the extracted data
stockprice_path = 'stock-tweet-and-price/stockprice/'  # Update this path as needed
stocktweet_path = 'stock-tweet-and-price/stocktweet/stocktweet.csv'  # Update this path as needed

# Connect to MySQL database
engine = create_engine('mysql+pymysql://root:your_password@localhost/stock_tweet_analysis')

# Step 1: Load stock prices CSVs into the database
# Iterate over each CSV file in the stockprice folder and load into MySQL
for file_name in os.listdir(stockprice_path):
    if file_name.endswith('.csv'):
        file_path = os.path.join(stockprice_path, file_name)
        stock_prices_df = pd.read_csv(file_path)
        
        # Load data into the stock_prices table
        stock_prices_df.to_sql('stock_prices', con=engine, if_exists='append', index=False)
        print(f"Loaded {file_name} into stock_prices table")

# Step 2: Load the tweets CSV into the database
tweets_df = pd.read_csv(stocktweet_path)
tweets_df.to_sql('tweets', con=engine, if_exists='append', index=False)
print("Loaded stocktweet.csv into tweets table")

# Optional: Query and explore data
stock_data = pd.read_sql('SELECT * FROM stock_prices LIMIT 5;', con=engine)
tweet_data = pd.read_sql('SELECT * FROM tweets LIMIT 5;', con=engine)

# Display the first few rows of data
print("Sample Stock Prices Data:", stock_data.head())
print("Sample Tweets Data:", tweet_data.head())

```
